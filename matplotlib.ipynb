```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t= np.linspace(0,40,1000) #start, finish, n points
#ei limit change kore kore dekhte hobe je kon limit e plotting better hoy
```

distance travelled by robber d=2.5t


```python
d_r=2.5*t
```

distance travelled by sheriff d= 3(t-5)


```python
d_s=3*(t-5)
```

```python
fig,ax= plt.subplots()
plt.title('A bank robber caught ')
plt.xlabel('time in mint')
plt.ylabel('distance in km')
ax.set_xlim([0,40]) #x axis borabor divisions
ax.set_ylim([0,100]) #y axis borabor divisions
ax.plot(t,d_r,c='green')
ax.plot(t,d_s,c='brown')
plt.axvline(x=30,color='purple',linestyle='--') #vertical line
plt.axhline(y=75,color='purple', linestyle='--')#horizontal line
# across the crossing point
```
